```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("treedata.csv")
```

Penjelasan data kolom dapat dilihat di : https://data.cityofnewyork.us/Environment/2015-Street-Tree-Census-Tree-Data/uvpi-gqnh

```python
data.head()
```

```python
#convert data type
data = data.convert_dtypes()
data["created_at"] = pd.to_datetime(data["created_at"])
data.info()
```

```python
#check apakah tree_id benar2 unik
len(data["tree_id"].unique()) == len(data)
```

```python
#Nilai tree_dbh dan stump_diam tidak boleh sama2 ada angka, salah satu harus 0 atau 2 2 nya 0, baca keterangan kolom untuk penjelasan.
data.loc[((data["stump_diam"] != 0) & (data["tree_dbh"] != 0)).to_numpy()]
```

```python
#check data validation apakah ada typo.
data["curb_loc"].value_counts()
```

```python
#check data validation apakah ada typo.
data["status"].value_counts()
```

```python
#membuang kolom yang tidak diperlukan, kita akan membuang kolom yang tidak diperlukan dan kolom semua detail lokasi dan memakai kolom "borough" saja
data.drop(data.loc[:,"cncldist":], axis=1,inplace=True)
data.drop(data.loc[:,"address":"community board"], axis=1, inplace=True)
data
```

```python
#checking apakah ada nilai null
data.isna().sum()
```

```python
#kebanyakan null values dari status tree "stump" ataupun "dead"
data.loc[(data['health'].isna()).to_numpy()].head(10)
```

```python
#replace null value dengan "not alive"
data.fillna("NotAlive", inplace=True)
```

```python
data["problems"].value_counts()
```

```python
#menghitung jumlah masalah dari satu pohon
data["ProblemsCount"] = data.apply(lambda x: (x.loc["root_stone":"brch_other"] == "Yes").sum(), axis=1)
```

```python
#mengecek konsistensi data
data.groupby('health')['ProblemsCount'].agg('mean')
```

Untuk memastikan konsistensi data, semakin bagus kesehatan pohon maka semakin rendah angka jumlah masalah.

```python
data.groupby(['curb_loc','status'])['ProblemsCount'].agg('mean')
```

bisa dilihat bahwa pohon yang berdiri diatas trotoar juga memiliki lebih banyak masalah dari pohon yang tidak berada ditrotoar. Note: tahap ini termasuk tahap data analisis dan bukan data validation dikarenakan tidak ada jaminan bahwa pohon yang berada ditrotoar selalu memiliki masalah lebih banyak dari yang tidak berada ditrotoar.
