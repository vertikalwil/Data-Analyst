```python
import pandas as pd
import numpy as np
from datetime import timedelta
```

```python
data = pd.read_csv("Data.csv", delimiter=';')
```

```python
data
```

```python
# membuang kolom yang tidak diperlukan
data.drop("Unnamed: 0", axis=1,inplace=True)
```

```python
# membuat index dari kolom keterangan nama, kolom ini akan dipakai sebagai acuan untuk meremove rows yang tidak diperlukan dan membuat data frame untuk nama bulan.
data.set_index('Unnamed: 1',drop=False, inplace=True)
```

```python
# membuat data frame untuk menyimpan nama bulan
datamonth = data.loc[data['Unnamed: 1'].isna().to_numpy()]
```

```python
#reset index dimulai dari angka 0 , index ini sebagai patokan untuk mencocokan tanggal dan bulan.
datamonth = datamonth["Jan-21"].reset_index(drop=True)
```

```python
datamonth
```

```python
#mengubah data type tanggal dari string menjadi date. tujuannya karena kedepannya kita akan melakukan kalkulasi menggunakan tanggal ini.
datamonth = pd.to_datetime(datamonth, format="%b-%y")
```

```python
datamonth
```

```python
#menambah bulan januari ke data frame, dikarenakan bulan januari ini terintegrasi kedalam kolom di data raw sehingga tidak terikut saat membuat dataframe ini.
datamonth[4] = pd.datetime(2021,1,1)
datamonth = datamonth.sort_values(ascending=True).reset_index(drop=True)
datamonth
```

```python
# mengubah nama kolom menjadi angka hari.
data.columns = np.arange(0,32)
# membuang baris yang berisi bulan karena bulan tersebut sudah kita input kedalam data frame 'datamonth'
data.drop([np.nan], axis=0, inplace=True)
```

```python
data
```

```python
#disini dibuat satu function yang menghasilkan integer untuk membedakan antara tanggal hari dengan bulan
a = -1
def indexes(x):
    global a
    if (x.name =="Team"):
        a += 1
        return a
    else:
        return a
data['indexes'] = data.apply(indexes, axis=1)
```

```python
#mengubah kolom indexes menjadi index
data = data.drop('Team', axis=0).set_index('indexes',drop=True)
data # untuk index 0 artinya bulan januari, untuk index 1 artinya bulan februari , dst.
```

```python
#unpivot table.
data = data.melt(id_vars = 0, ignore_index=False)
data
```

```python
# mengubah data type int variable menjadi timedelta untuk melakukan perhitungan dengan data frame datamonth. variable disini adalah jumlah hari.
data['variable'] = data.apply(lambda x: pd.to_timedelta(x['variable'], unit='d'), axis = 1) 
```

```python
# assign variable dengan nilai baru dalam bentuk datetime
data['variable'] = datamonth.loc[data.index] + data['variable'] - pd.to_timedelta(1, unit='d')
```

```python
#ubah nama kolom, sorting value dan set index untuk menghilangkan nan value.
data.columns = ["Nama","Tanggal","Kehadiran"]
data.sort_values("Tanggal", inplace=True)
data.set_index("Kehadiran",drop = False, inplace=True)
```

```python
#membuang kehadiran yang NaN, value Nan bisa terjadi karena kebetulan di hari itu adalah hari libur atau hari tersebut tidak ada dalam bulan tertentu.
data.drop(np.nan , axis=0, inplace=True)
```

```python
#hasil akhir
data.reset_index(drop=True,inplace=True)
data
```
